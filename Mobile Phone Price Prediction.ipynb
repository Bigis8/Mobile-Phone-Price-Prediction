# BUILDING THE MODEL

```python
X=train_file[[ "battery_power", "blue", "clock_speed", "dual_sim", "fc","four_g", "int_memory", "m_dep", "mobile_wt",
      "n_cores", "px_height", "px_width", "ram", "sc_h", "sc_w", "talk_time","three_g","touch_screen", "wifi"  ]]

Y= train_file["price_range"]
```

```python
from sklearn.preprocessing import MinMaxScaler

Min =MinMaxScaler()
train_file = Min.fit_transform(train_file)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,   y_train,y_test = train_test_split(train_file,Y, test_size=0.2, random_state=40)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
model_one = dt.fit(x_train, y_train)
```

```python
y_predict=model_one.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,y_predict )
```
